```python
#list files in a directory
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.to_csv('all_data.csv', index=False)
```

#### Read in updated Dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

#### CleanUp the Data

```python
#check the rows with NAN values
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
#drop NAN values
all_data = all_data.dropna()
all_data.head()
```

```python
#### Find 'Or' and delete it
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']

                   
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
```

```python
all_data.head()
```

# What month had the highest sales?

```python
#### Adding a Sales Column
#view data types.
all_data.dtypes
```

```python
#convert columns to the right types
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
# group sales by month
all_data.groupby('Month').sum()
```

```python
# plot data
import matplotlib.pyplot as plt

sales_by_month = all_data.groupby('Month').sum()
months = range(1,13)

plt.bar(months, sales_by_month['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month NUmber')
plt.show()
```

# What city had the highest number of sales?

```python
# What city had the highest number of sales.
#group sales by city
#first we need to extract the city from the address
all_data['State'] = all_data['Purchase Address'].str[-8:-6]
all_data.head()
```

```python
# Add column for City
#convert the column to string
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
#Plot data

city = [city for city, df in all_data.groupby(['City'])]

plt.bar(city, results['Sales'])
plt.xticks(city, rotation = 'vertical', size = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

# What time should we display advertisement to maximize likelihood of  customer's buying the product?

```python
all_data.head()
```

```python
#convert order data to datetime
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
#add hour couln
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
#Plot data

hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks = hours
plt.grid()
plt.xlabel('Hours')
plt.ylabel('Number of Hours')
plt.show()
```

# What products are most often sold together?

```python
all_data.head()
```

```python
all_data['Product'].describe()
```

```python
all_data['Product'].unique
```

```python
result = all_data['Product'].groupby(all_data['Order ID'])
result
```

```python
tdf = all_data[all_data['Order ID'].duplicated(keep=False)]
```

```python
# group the items
tdf['Grouped'] = tdf.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

tdf = tdf[['Order ID', 'Grouped']].drop_duplicates()

tdf.head()
```

```python
from itertools import combinations
from collections import Counter


count = Counter()

for row in tdf['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

```python
# need to plot pproduct against sales
import matplotlib.pyplot as plt
product_group = all_data.groupby('Product')

quantity = product_group.sum()['Quantity Ordered']

product = [product for product, df in product_group]

plt.bar(product , quantity)

plt.xticks(product, rotation='vertical',size=8)

plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(product , quantity, color = 'g')
ax2.plot(product, prices, 'b-')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Price ($)', color = 'b')
ax1.set_xticklabels(product, rotation='vertical',size=8)

plt.show()
```
